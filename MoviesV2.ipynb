```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np

import math
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots we will create

# Read in the data

df = pd.read_csv(r'/Users/michaelneumann/Downloads/movies.csv')

```

```python
# Let's Look at the data

df.head()
```

```python
# Let's look for missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, pct_missing))
```

```python
# removing missing data
df = df.dropna()
```

```python
# Data Types for our columns

df.dtypes
```

```python
# Updating data types

df['budget'] = df['budget'].astype('int64')
```

```python
df['gross'] = df['gross'].astype('int64')
df['votes'] = df['votes'].astype('int64')
```

```python
df.head()
```

```python
#Creating Correct Year Column

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)

df
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any Duplicates

df['company'].drop_duplicates().sort_values(ascending=False)
#df.drop_duplicates()
```

```python
# Theory - Budget will be highly correlated to revenue
# Company will have high correlation
# Scatter plot Budget vs Gross

plt.scatter(x=df['budget'], y=df['gross'])


plt.title('Budget vs Gross Earnings')

plt.ylabel('Gross Earnings')

plt.xlabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# Plot budget vs Gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color":"blue"})
```

```python
#Let's start looking at correlation
# Methods - Pearson, Kendall, Spearman

df.corr(method='pearson') 
```

```python
# High correlation between budget and gross was accurate

correlation_matrix = df.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Movie Attributes')

plt.ylabel('Movie Attributes')

plt.xlabel('Movie Attributes')

plt.show()
```

```python
# Look at Company

df.head()
```

```python
df_numerized = df.copy()

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matric for Movie Attributes')

plt.ylabel('Movie Attributes')

plt.xlabel('Movie Attributes')

plt.show()
```

```python
df_numerized.corr()
```

```python
correlation_mat = df_numerized.corr()

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()

sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
